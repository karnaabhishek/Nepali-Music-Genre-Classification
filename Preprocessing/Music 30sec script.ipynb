```python
import os
from mutagen.mp3 import MP3
from pydub import AudioSegment
```

```python
inputdir="/content/drive/MyDrive/Songs Collection/lok dohori"
```

```python
for filename in os.listdir(inputdir):
  save_file_name = filename[:-4]
  print(save_file_name)
  myaudio = AudioSegment.from_file(inputdir+"/"+filename, "mp3")
  audio = MP3(inputdir+"/"+filename)
  length=audio.info.length
  length=length*1000
  start=0
  end=30000
  i=1
  while (end<length):
    clip=myaudio[start:end]
    clip.export("/content/drive/MyDrive/songs/lok dohori/"+save_file_name+"_"+str(i)+".wav", format="wav")
    start+=30000
    end+=30000
    i+=1
```
